```python
import numpy as np
from rdkit import DataStructs
def mol2fp(mol):
    fp = AllChem.GetHashedMorganFingerprint(mol, 2, nBits=1024)
    ar = np.zeros((1,), dtype=np.int8)
    DataStructs.ConvertToNumpyArray(fp, ar)
    return ar
```

```python
Catmos_df["FPs"] = Catmos_df.Molecule.apply(mol2fp)
Catmos_df.head(2)
```

```python
def CalLogLD50(LD50):
    LogLD50=np.log10(LD50)
    return LogLD50
```

```python
Catmos_df["LogLD50"] = Catmos_df["CATMoS_LD50_mgkg"].apply(CalLogLD50)
```

```python
def MACCSfp(mol):
    fp = rdkit.Chem.rdMolDescriptors.GetMACCSKeysFingerprint(mol)
    ar = np.zeros((1,), dtype=np.int8)
    DataStructs.ConvertToNumpyArray(fp, ar)
    return ar
```

```python
Catmos_df["MACCSFPs"] = Catmos_df.Molecule.apply(MACCSfp)
Catmos_df.head(2)
```
